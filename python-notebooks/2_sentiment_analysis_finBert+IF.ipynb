## Pre-trained Distilled BERT for Financial News Sentimnt Analysis

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

# Load model directly
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis")
model = AutoModelForSequenceClassification.from_pretrained("mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis")

sentiment_analysis = pipeline("sentiment-analysis",model="mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis")
```

```python
df = pd.read_csv('tesla_financial_news_january.csv', parse_dates=['Date'])
df_copy = df.copy()

df_copy.drop(columns=['Content','Source','URL'],inplace=True)
df_copy.drop_duplicates(subset=['Title'],inplace=True)


df_copy['Title'] = df_copy['Title'].astype('string')

average_df = df_copy.groupby('Date')['Title']
average_df.describe()
```

```python
def perform_sentiment_news():
    for index, row in df_copy.iterrows():
        sentiment = sentiment_analysis(row['Title'])
        sentiment_object = sentiment.pop()
        df_copy.at[index, 'Sentiment'] = sentiment_object['label']
        df_copy.at[index, 'Sentiment Score'] = sentiment_object['score']
```

```python
#calculate daily sentiment
def calculate_daily_weighted_sentiment(x):
    
    pos_score = x.query("Sentiment == 'positive'")['Sentiment Score'].sum()
    neg_score = x.query("Sentiment == 'negative'")['Sentiment Score'].sum()
    neu_score = x.query("Sentiment == 'neutral'")['Sentiment Score'].sum()

    
    if (pos_score + neg_score) != 0:
        news_weigthed = (pos_score - neg_score)/(pos_score + neg_score)
    else:
        news_weigthed = pos_score
    
    x['Daily Weighted Sentiment'] = news_weigthed
    
    return x
```

```python
news_sentiments = perform_sentiment_news()
df_copy['Sentiment'] = df_copy['Sentiment'].astype('string')

#calculate daily sentiment frequency
df_daily_sentiment = df_copy.groupby(['Date']).apply(calculate_daily_weighted_sentiment)
df_daily_sentiment.drop(columns=['Date'], inplace=True)
```

# Stock Price features engineering 

```python
df_stock_price = pd.read_csv('Tesla (TSLA) Historical Data.csv', parse_dates=['Date'], index_col='Date')
df_stock_price['Change %'] = df_stock_price['Change %'].map(lambda x: x.rstrip('%'))
df_stock_price['Vol.'] = df_stock_price['Vol.'].map(lambda x: x.rstrip('M'))

df_stock_price['Vol.']=pd.to_numeric(df_stock_price['Vol.'])
df_stock_price['Change %']=pd.to_numeric(df_stock_price['Change %'])
df_stock_price
```

```python
df_combined_data= pd.merge(df_daily_sentiment, df_stock_price, on='Date')
df_combined_data.reset_index(inplace=True)
df_combined_data.drop(columns=['Title','Sentiment','Sentiment Score','Open','High','Low'], inplace=True)
df_combined_data.drop_duplicates(inplace=True)
df_combined_data
```

## Check Correlations between sentiment, change % and Price

```python
#df_normalized_combined
corr_mtx_after_norm = df_combined_data.corr()
#corr_mtx_after_norm = df_stock_price.corr()
# Visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_mtx_after_norm, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

## Isolation Forest for Anomaly Detection

```python
# --- Anomaly Detection using Isolation Forest ---
features = ['Daily Weighted Sentiment','Change %']  # Features used for anomaly detection
X = df_combined_data[features]
# Train Isolation Forest model
model = IsolationForest(max_samples='auto', contamination=0.01)  # Adjust contamination based on expected anomaly ratio
df_combined_data['Anomaly'] = model.fit_predict(X)

# Anomalies are marked as -1
df_combined_data['Anomaly_Label'] = df_combined_data['Anomaly'].apply(lambda x: 'Disruptive' if x == -1 else 'Not Disruptive')

# Check results
print(df_combined_data[['Daily Weighted Sentiment', 'Price', 'Vol.', 'Change %', 'Anomaly_Label']])
```

## Plot Result Isolation Forest

```python

# --- Plotting Stock Prices, Sentiment, and Disruptive Events ---
fig, ax1 = plt.subplots(figsize=(25, 10))

# Plot Stock Price on the first y-axis (Blue)
ax1.set_xlabel("Date")
ax1.set_ylabel("Stock Close Price", color="blue")
ax1.plot(df_combined_data["Date"], df_combined_data["Change %"], color="blue",  label="Stock Close Price")

# Create a second y-axis for the sentiment scores (Green)
ax2 = ax1.twinx()
ax2.set_ylabel("Sentiment Score", color="green")
ax2.plot(df_combined_data["Date"], df_combined_data["Daily Weighted Sentiment"], color="green", linestyle="dashed", label="Daily Weighted Sentiment")
ax2.tick_params(axis="y", labelcolor="green")

# Highlight disruptive events with red dots
disruptive = df_combined_data[df_combined_data['Anomaly_Label'] == 'Disruptive']
ax1.scatter(disruptive['Date'], disruptive['Change %'], color='red', label='Disruptive Events')

# Add title and labels
fig.tight_layout()
plt.title("Stock Prices, Sentiment Trends, and Disruptive Events Over Time")

# Show legends for both axes
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

ax1.xaxis.set_major_locator(mdates.DayLocator(interval=1))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.xticks(rotation=45)
plt.show()
```
